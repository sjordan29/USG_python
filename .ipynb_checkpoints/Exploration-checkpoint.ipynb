```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
init = pd.read_csv('init_cells.csv')
init.head()
```

Plot the index number of voronoi region associated with each node. The nodes here are the center points of a cell. 

Read more about Voronoi diagrams at [wolfram alpha](https://mathworld.wolfram.com/VoronoiDiagram.html).

```python
fig, ax = plt.subplots()
points = ax.scatter(init.x, init.y, s=25, c=init.vor_region, cmap='plasma')
fig.colorbar(points)
plt.show()
```

HEC-RAS makes a computational mesh by following the Delaunay Triangulation technique and then constructing a Voronoi diagram ([see documentation](https://www.hec.usace.army.mil/software/hec-ras/documentation/HEC-RAS%205.0%202D%20Modeling%20Users%20Manual.pdf)).
